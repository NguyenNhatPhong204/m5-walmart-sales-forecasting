```python
import pandas as pd
import numpy as np

def reduce_mem_usage(df):
    """
    Hàm lặp qua tất cả các cột của DataFrame và thay đổi kiểu dữ liệu 
    để giảm dung lượng bộ nhớ.
    """
    start_mem = df.memory_usage().sum() / 1024**2
    print(f'Dung lượng ban đầu của DataFrame là {start_mem:.2f} MB')
    
    for col in df.columns:
        col_type = df[col].dtype
        
        # Bỏ qua các cột chứa ngày tháng đã được parse (nếu có)
        if col_type != object and col_type.name != 'category' and 'datetime' not in col_type.name:
            c_min = df[col].min()
            c_max = df[col].max()
            
            # Xử lý dữ liệu số nguyên
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            # Xử lý dữ liệu số thực
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        # Xử lý chuỗi ký tự (chuyển sang category)
        elif col_type == object:
            df[col] = df[col].astype('category')

    end_mem = df.memory_usage().sum() / 1024**2
    print(f'Dung lượng sau khi tối ưu là: {end_mem:.2f} MB')
    print(f'Đã giảm được {100 * (start_mem - end_mem) / start_mem:.1f}%\n')
    
    return df

# Đọc dữ liệu thô
print("--- Tải và tối ưu dữ liệu Sales ---")
sales = pd.read_csv('../data_raw/sales_train_evaluation.csv')
sales = reduce_mem_usage(sales)

print("--- Tải và tối ưu dữ liệu Calendar ---")
calendar = pd.read_csv('../data_raw/calendar.csv')
calendar = reduce_mem_usage(calendar)

print("--- Tải và tối ưu dữ liệu Prices ---")
prices = pd.read_csv('../data_raw/sell_prices.csv')
prices = reduce_mem_usage(prices)
```
